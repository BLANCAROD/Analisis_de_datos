```python
#Se pueden visualizar algunas características del dataset. Por ejemplo: histograma del Sexo de los pasajeros.
```

```python
sf["Sex"].hist(bins=30)
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
sf["Age"].hist(bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
#7.	Se crean diagramas de barras para mostrar cantidad de pasajeros sobrevivientes y no sobrevivientes, 
#así como los pasajeros de sexo masculino y los de sexo femenino.
```

```python
survived_counts=sf["Survived"].value_counts()
plt.bar(survived_counts.index, survived_counts.values)
plt.xticks([0, 1], ["Not Survived", "Survived"])
plt.ylabel("Count")
plt.show()
```

```python
survived_counts=sf["Sex"].value_counts()
plt.bar(survived_counts.index, survived_counts.values)
plt.xticks([0, 1], ["Masculino", "Femenino"])
plt.ylabel("Count")
plt.show()
```
